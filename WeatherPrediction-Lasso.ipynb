```python
#independent and dependent features
X=df.iloc[:,[1,2,5,6,7,8,9]]
Y=df.iloc[:,3]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.33)
```

```python
X_test.shape
```

```python
#Lasso
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

las=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,2,1,5,10,20,30,35,40,45,50,55,100]}
las_Regressor = GridSearchCV(las,parameters,scoring='r2',cv=5)
las_Regressor.fit(X,Y)
```

```python
las_Regressor.score(X_train,y_train)
```

```python
las_Regressor.score(X_test,y_test)
```

```python
print(las_Regressor.best_params_)
print(las_Regressor.best_score_)
```

```python
prediction_lasso = las_Regressor.predict(X_test)
```

```python
sns.distplot(y_test-prediction_lasso)
```

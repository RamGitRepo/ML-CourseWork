```python
df=df.drop(['Formatted Date','Precip Type','Apparent Temperature (C)','Humidity','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)','Pressure (millibars)','Daily Summary'],axis='columns')
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
df.iloc[:,0]=enc.fit_transform(df.iloc[:,0])
```

```python
df.head()
```

```python
plt.scatter(df['Temperature (C)'],df['Summary'])
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
```

```python
km
```

```python
y_predict = km.fit_predict(df[['Temperature (C)','Summary']])
```

```python
y_predict
```

```python
df['cluster'] = y_predict
```

```python
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]


plt.scatter(df1['Temperature (C)'],df1['Summary'], color='green')
plt.scatter(df2['Temperature (C)'],df2['Summary'], color='blue')
plt.scatter(df3['Temperature (C)'],df3['Summary'], color='black')

plt.xlabel('Temperature (C)')
plt.ylabel('Summary')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[['Temperature (C)']])
df[['Temperature (C)']] = scaler.transform(df[['Temperature (C)']])
df
```

```python
scaler.fit(df[['Summary']])
df[['Summary']] = scaler.transform(df[['Summary']])
df
```

```python
km = KMeans(n_clusters=3)
y_predict = km.fit_predict(df[['Temperature (C)','Summary']])
```

```python
df['scaleCluster'] = y_predict
```

```python
df
```

```python
df1 = df[df.scaleCluster==0]
df2 = df[df.scaleCluster==1]
df3 = df[df.scaleCluster==2]


plt.scatter(df1['Temperature (C)'],df1['Summary'], color='green')
plt.scatter(df2['Temperature (C)'],df2['Summary'], color='blue')
plt.scatter(df3['Temperature (C)'],df3['Summary'], color='black')
plt.legend()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.scaleCluster==0]
df2 = df[df.scaleCluster==1]
df3 = df[df.scaleCluster==2]

plt.scatter(df1['Temperature (C)'],df1['Summary'], color='green')
plt.scatter(df2['Temperature (C)'],df2['Summary'], color='blue')
plt.scatter(df3['Temperature (C)'],df3['Summary'], color='black')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], color='red', marker='*', label="centroid")
plt.legend()
```

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['Temperature (C)','Summary']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K-Value')
plt.ylabel('Sum of Squared Error')
plt.plot(k_range,sse)
```

```python
km.labels_
```

```python
clusterer = KMeans(n_clusters=3,random_state=10)
cluster_label = clusterer.fit_predict(km.cluster_centers_)
```
